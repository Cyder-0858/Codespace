## Â¿QuÃ© es un montÃ­culo?

Un montÃ­culo es una estructura de datos basada en un Ã¡rbol binario completo que permite acceder rÃ¡pidamente al elemento de mayor (o menor) prioridad.

### Tipos de montÃ­culos:
- **MontÃ­culo mÃ¡ximo**: el valor de cada nodo es mayor o igual que el de sus hijos.
- **MontÃ­culo mÃ­nimo**: el valor de cada nodo es menor o igual que el de sus hijos.

### RepresentaciÃ³n en listas:
- Padre de `i`: `(i - 1) // 2`
- Hijo izquierdo de `i`: `2 * i + 1`
- Hijo derecho de `i`: `2 * i + 2`

```python
class Monticulo:
    def __init__(self):
        self.datos = []

    @staticmethod
    def generar(lista):
        """Genera un montÃ­culo a partir de una lista desordenada."""
        monticulo = Monticulo()
        for elemento in lista:
            monticulo.insertar(elemento)
        return monticulo

    def insertar(self, elemento):
        """Inserta un elemento y lo hace flotar para mantener la propiedad del montÃ­culo."""
        self.datos.append(elemento)
        self._flotar(len(self.datos) - 1)

    def _flotar(self, i):
        """
        Hace flotar el elemento en la posiciÃ³n i si es mayor que su padre.
        
        COMPARACIÃ“N DE TUPLAS PARA MANTENER FIFO:
        =========================================
        Cuando trabajamos con ColaPrioridad, self.datos[i] contiene tuplas: (prioridad, -contador, dato)
        
        La comparaciÃ³n self.datos[i] > self.datos[(i - 1) // 2] funciona asÃ­:
        1. Primero compara prioridades: si 3 > 2, entonces (3, -x, dato) > (2, -y, dato)
        2. Si las prioridades son iguales, compara contadores: si -0 > -1, entonces sale primero
        3. Python hace esta comparaciÃ³n lexicogrÃ¡fica automÃ¡ticamente
        
        Ejemplo: 
        - Elemento hijo: (2, -0, "Primera tarea prioridad 2")    # LlegÃ³ primero
        - Elemento padre: (2, -1, "Segunda tarea prioridad 2")   # LlegÃ³ despuÃ©s
        - ComparaciÃ³n: (2, -0, ...) > (2, -1, ...) â†’ True, por lo que flota el hijo
        """
        while i > 0 and self.datos[i] > self.datos[(i - 1) // 2]:
            self.datos[i], self.datos[(i - 1) // 2] = self.datos[(i - 1) // 2], self.datos[i]
            i = (i - 1) // 2

    def borrar(self):
        """Elimina y devuelve el elemento de mayor prioridad (la raÃ­z)."""
        if self.vacio():
            return None
        raiz = self.datos[0]
        ultimo = self.datos.pop()
        if not self.vacio():
            self.datos[0] = ultimo
            self._hundir(0)
        return raiz

    def _hundir(self, i):
        """
        Hace hundir el elemento en la posiciÃ³n i si es menor que alguno de sus hijos.
        
        ðŸ” COMPARACIÃ“N DE TUPLAS DURANTE HUNDIMIENTO:
        =============================================
        Durante el hundimiento, las comparaciones automÃ¡ticas de tuplas mantienen
        tanto el orden de prioridades como el comportamiento FIFO.
        
        CÃ“MO FUNCIONA LA COMPARACIÃ“N AUTOMÃTICA:
        1. Python ejecuta: self.datos[hijo_izq] > self.datos[mayor]
        2. Compara lexicogrÃ¡ficamente: (prioridad1, contador1, dato1) vs (prioridad2, contador2, dato2)
        3. Si prioridad1 â‰  prioridad2 â†’ gana la mayor prioridad
        4. Si prioridad1 = prioridad2 â†’ compara contadores (el menor contador negativo gana)
        5. Â¡Python maneja todo automÃ¡ticamente!
        
        EJEMPLO PASO A PASO:
        ===================
        Estado inicial despuÃ©s de borrar raÃ­z:
        - Padre (nuevo en raÃ­z): (2, -5, "Sexta tarea")    # LlegÃ³ sexto, prioridad 2
        - Hijo izquierdo:        (2, -3, "Cuarta tarea")   # LlegÃ³ cuarto, prioridad 2
        - Hijo derecho:          (1, -7, "Octava tarea")   # LlegÃ³ octavo, prioridad 1
        
        Comparaciones automÃ¡ticas:
        1. Â¿(2, -3, ...) > (2, -5, ...)? â†’ SÃ (-3 > -5, llegÃ³ antes)
        2. Â¿(1, -7, ...) > (2, -5, ...)? â†’ NO (1 < 2, menor prioridad)
        
        Resultado: hijo_izq se convierte en "mayor", se intercambia con padre.
        Â¡El elemento que llegÃ³ antes (cuarto) sube sobre el que llegÃ³ despuÃ©s (sexto)!
        
        MANTENIMIENTO AUTOMÃTICO DEL ORDEN FIFO:
        ========================================
        - Las tuplas garantizan que elementos con igual prioridad se ordenen por llegada
        - Python's comparaciÃ³n lexicogrÃ¡fica hace todo el trabajo pesado
        - No necesitamos lÃ³gica especial para FIFO, Â¡estÃ¡ integrada en la estructura!
        """
        tamaÃ±o = len(self.datos)
        while True:
            hijo_izq = 2 * i + 1
            hijo_der = 2 * i + 2
            mayor = i  # Inicializar mayor con i

            # Comparar con hijo izquierdo (lexicogrÃ¡ficamente si son tuplas)
            if hijo_izq < tamaÃ±o and self.datos[hijo_izq] > self.datos[mayor]:
                mayor = hijo_izq
            # Comparar con hijo derecho (lexicogrÃ¡ficamente si son tuplas)    
            if hijo_der < tamaÃ±o and self.datos[hijo_der] > self.datos[mayor]:
                mayor = hijo_der
            # Si no hay intercambio necesario, terminar
            if mayor == i:
                break
            # Intercambiar elementos manteniendo orden FIFO
            self.datos[i], self.datos[mayor] = self.datos[mayor], self.datos[i]
            i = mayor

    def vacio(self):
        """Devuelve True si el montÃ­culo estÃ¡ vacÃ­o."""
        return len(self.datos) == 0

    @staticmethod
    def ordena(lista):
        """Ordena una lista usando un montÃ­culo."""
        monticulo = Monticulo.generar(lista)
        resultado = []
        while not monticulo.vacio():
            resultado.insert(0, monticulo.borrar())
        return resultado
```

##Colas de prioridad
Una cola de prioridad es una estructura FIFO, donde los elementos tienen asociada una prioridad de forma que el primero serÃ¡ el primero que llegÃ³ con mÃ¡xima prioridad.

```python
class ColaPrioridad:
    def __init__(self):
        self.monticulo = Monticulo()
        self.contador = 0  # Contador global para garantizar orden FIFO en igual prioridad

    def agregar(self, prioridad, dato):
        """
        Agrega un dato con su prioridad a la cola.
        
        MECANISMO FIFO PARA IGUAL PRIORIDAD:
        ====================================
        Para mantener el orden de llegada (FIFO) cuando hay elementos con la misma prioridad,
        utilizamos una tupla de 3 elementos: (prioridad, -contador, dato)
        
        Â¿Por quÃ© funciona este mecanismo?
        1. PRIORIDAD (primer elemento): Determina quÃ© elementos se atienden primero
        2. -CONTADOR (segundo elemento): ActÃºa como "desempate" para igual prioridad
           - Se usa NEGATIVO (-self.contador) porque el montÃ­culo es MÃXIMO
           - El elemento que llegÃ³ PRIMERO tiene contador mÃ¡s BAJO
           - Al negarlo, se convierte en valor mÃ¡s ALTO en el montÃ­culo
           - Ejemplo: primer elemento contador=0 â†’ -0=0 (mayor)
                     segundo elemento contador=1 â†’ -1=-1 (menor)
        3. DATO (tercer elemento): La informaciÃ³n real que queremos almacenar
        
        Ejemplo de funcionamiento:
        - Agregar("A", prioridad=2) â†’ tupla: (2, -0, "A")
        - Agregar("B", prioridad=2) â†’ tupla: (2, -1, "B")  
        - Agregar("C", prioridad=2) â†’ tupla: (2, -2, "C")
        
        En el montÃ­culo mÃ¡ximo:
        - Primero se comparan prioridades: 2 = 2 = 2 (empate)
        - Luego se comparan contadores: 0 > -1 > -2
        - Por tanto, el orden de salida serÃ¡: A, B, C (FIFO)
        """
        # Crear tupla con prioridad, contador negativo y dato
        self.monticulo.insertar((prioridad, -self.contador, dato))
        self.contador += 1  # Incrementar contador para prÃ³ximo elemento

    def atender(self):
        """
        Atiende el elemento de mayor prioridad respetando FIFO para igual prioridad.
        
        PROCESO DE ELIMINACIÃ“N:
        ======================
        1. El montÃ­culo automÃ¡ticamente devuelve la tupla con mayor valor
        2. La comparaciÃ³n de tuplas en Python es lexicogrÃ¡fica:
           - Primero compara elemento [0] (prioridad)
           - Si hay empate, compara elemento [1] (contador negativo)
           - El contador negativo asegura que el primer elemento agregado salga primero
        3. Extraemos solo el dato (Ã­ndice 2) de la tupla resultado
        
        Ejemplo de extracciÃ³n:
        - MontÃ­culo contiene: [(3, -0, "Tarea Snoke"), (2, -0, "Tarea Phasma")]  
        - Se extrae: (3, -0, "Tarea Snoke")
        - Se retorna: "Tarea Snoke"
        """
        if self.monticulo.vacio():
            return None
        resultado = self.monticulo.borrar()
        return resultado[2]  # Retornar solo el dato (Ã­ndice 2) de la tupla
    
    def vacia(self):
        """Devuelve True si la cola estÃ¡ vacÃ­a."""
        return self.monticulo.vacio()
```

```python
"""
ðŸ” EXPLICACIÃ“N DETALLADA: COMPARACIÃ“N DE TUPLAS EN PYTHON
=========================================================

La clave del funcionamiento FIFO en nuestra cola de prioridad estÃ¡ en cÃ³mo
Python compara automÃ¡ticamente las tuplas. Â¡Veamos exactamente cÃ³mo funciona!

COMPARACIÃ“N LEXICOGRÃFICA PASO A PASO:
======================================
Python compara tuplas elemento por elemento, similar a cÃ³mo se ordenan
las palabras en un diccionario (de ahÃ­ "lexicogrÃ¡fica").

ALGORITMO DE COMPARACIÃ“N DE TUPLAS:
1. Compara el primer elemento de ambas tuplas
2. Si son diferentes â†’ la tupla con mayor primer elemento es "mayor"
3. Si son iguales â†’ compara el segundo elemento
4. Si siguen iguales â†’ compara el tercer elemento
5. ContinÃºa hasta encontrar diferencia o agotar elementos

EJEMPLOS REALES CON NUESTRAS TUPLAS:
===================================
"""

# Simulemos las tuplas que usa nuestra cola de prioridad
print("COMPARACIÃ“N DE TUPLAS - EJEMPLOS INTERACTIVOS")
print("=" * 50)

# Caso 1: Diferentes prioridades
tupla_snoke = (3, -0, "RevisiÃ³n sistemas defensa")
tupla_phasma = (2, -0, "Entrenamiento escuadra")

print("CASO 1 - Diferentes prioridades:")
print(f"Snoke:  {tupla_snoke}")
print(f"Phasma: {tupla_phasma}")
print(f"Â¿tupla_snoke > tupla_phasma? {tupla_snoke > tupla_phasma}")
print("RazÃ³n: 3 > 2 en el primer elemento, Â¡no necesita comparar mÃ¡s!")
print()

# Caso 2: Igual prioridad, diferente orden de llegada
tupla_phasma1 = (2, -0, "Primera tarea Phasma")    # contador=0 (llegÃ³ primero)
tupla_phasma2 = (2, -1, "Segunda tarea Phasma")    # contador=1 (llegÃ³ despuÃ©s)

print("CASO 2 - Igual prioridad, diferente orden:")
print(f"Primera: {tupla_phasma1}")
print(f"Segunda: {tupla_phasma2}")
print(f"Â¿primera > segunda? {tupla_phasma1 > tupla_phasma2}")
print("RazÃ³n: 2 = 2 (empate), pero -0 > -1, Â¡la primera tiene prioridad!")
print()

# Caso 3: DemostraciÃ³n del por quÃ© usar contador negativo
print("CASO 3 - Â¿Por quÃ© contador NEGATIVO?")
contador_positivo_1 = (2, 0, "Primera")   # Si usÃ¡ramos contador positivo
contador_positivo_2 = (2, 1, "Segunda")

print(f"Con contador positivo - Primera: {contador_positivo_1}")
print(f"Con contador positivo - Segunda: {contador_positivo_2}")
print(f"Â¿primera > segunda? {contador_positivo_1 > contador_positivo_2}")
print("Â¡ERROR! La segunda saldrÃ­a antes (1 > 0), rompiendo FIFO")
print()

print(f"Con contador NEGATIVO - Primera: {tupla_phasma1}")
print(f"Con contador NEGATIVO - Segunda: {tupla_phasma2}")
print(f"Â¿primera > segunda? {tupla_phasma1 > tupla_phasma2}")
print("Â¡CORRECTO! La primera sale antes (-0 > -1), manteniendo FIFO")
print()

# DemostraciÃ³n en orden de montÃ­culo
print("ORDEN EN MONTÃCULO MÃXIMO:")
tuplas = [
    (3, -0, "Snoke - Primera orden"),
    (3, -1, "Kylo Ren - Segunda orden"),
    (2, -2, "Phasma - Tercera orden"),
    (2, -3, "Phasma - Cuarta orden"),
    (1, -4, "Hux - Quinta orden")
]

tuplas_ordenadas = sorted(tuplas, reverse=True)  # Orden descendente (mÃ¡ximo)
print("Orden de salida del montÃ­culo:")
for i, tupla in enumerate(tuplas_ordenadas, 1):
    print(f"{i}Âº: {tupla}")

print("\nÂ¡FIFO mantenido dentro de cada prioridad!")
```

## Ejercicio prÃ¡ctoco: base Starkiller
3.	El general Hux es la persona encargada de administrar todas las operaciones de la base Starki- ller, para lo cual nos solicita desarrollar un algoritmo que permita controlar las actividades que se realizan, contemplando lo siguiente:

a.	debe contemplar distintas prioridades para el control de operaciones de acuerdo al siguien- te criterio: pedidos de lÃ­der supremo Snoke y de Kylo Ren nivel tres, de capitÃ¡n Phasma nivel dos y el resto de las operaciones nivel a cargo de los generales de la base de nivel uno;

b.	de cada actividad se conoce quien es el encargado, una descripciÃ³n, la hora y opcional- mente la cantidad de Stormtroopers requeridos;

c.	utilizar una cola de prioridad para administrar las distintas operaciones, debe cargar al menos ocho: dos de nivel tres, cuatro de nivel dos y cuatro de nivel uno;

d.	opcionalmente se podrÃ¡n agregar operaciones luego de atender una;

e.	realizar la atenciÃ³n de las operaciones de la cola;

f.	luego de atender la quinta operaciÃ³n, agregar una operaciÃ³n solicitada por capitÃ¡n Phasma para revisiÃ³n de intrusos en el hangar B7 que requiere 25 Stormstroopers;

g.	luego de atender la sexta operaciÃ³n, agregar una operaciÃ³n solicitada por el lÃ­der supremo Snoke para destruir el planeta Takodana.


```python
class Operacion:
    def __init__(self, encargado, descripcion, hora, stormtroopers=None):
        self.encargado = encargado
        self.descripcion = descripcion
        self.hora = hora
        self.stormtroopers = stormtroopers

    def __str__(self):
        texto = f"{self.descripcion} (Encargado: {self.encargado}, Hora: {self.hora})"
        if self.stormtroopers:
            texto += f", Stormtroopers requeridos: {self.stormtroopers}"
        return texto
```

```python
# Creamos operaciones como instancias de la clase Operacion
operaciones = [
    (3, Operacion("Snoke", "RevisiÃ³n de sistemas de defensa", "08:00")),
    (3, Operacion("Kylo Ren", "InspecciÃ³n de tropas", "09:00")),
    (2, Operacion("Phasma", "Entrenamiento de escuadra", "10:00")),
    (2, Operacion("Phasma", "RevisiÃ³n de armamento", "11:00")),
    (2, Operacion("Phasma", "SupervisiÃ³n de patrullas", "12:00")),
    (2, Operacion("Phasma", "Control de accesos", "13:00")),
    (1, Operacion("General Hux", "Informe diario", "14:00")),
    (1, Operacion("General Hux", "RevisiÃ³n de suministros", "15:00")),
    (1, Operacion("General Hux", "PlanificaciÃ³n de misiones", "16:00")),
    (1, Operacion("General Hux", "EvaluaciÃ³n de personal", "17:00")),
]

# Creamos la cola de prioridad
cola = ColaPrioridad()

# Cargamos las operaciones
for prioridad, op in operaciones:
    cola.agregar(prioridad, op)

print("=== Atendiendo operaciones ===")
# Atendemos operaciones y agregamos nuevas segÃºn el enunciado
contador_operaciones = 0
while not cola.vacia():
    op = cola.atender()
    contador_operaciones += 1
    print(f"OperaciÃ³n {contador_operaciones}: {op}")

    if contador_operaciones == 5:
        nueva_op = Operacion("Phasma", "RevisiÃ³n de intrusos en el hangar B7", "18:00", stormtroopers=25)
        cola.agregar(2, nueva_op)
        print("  -> Agregada: RevisiÃ³n de intrusos en el hangar B7")
    
    if contador_operaciones == 6:
        nueva_op = Operacion("Snoke", "DestrucciÃ³n del planeta Takodana", "19:00")
        cola.agregar(3, nueva_op)
        print("  -> Agregada: DestrucciÃ³n del planeta Takodana")
```
